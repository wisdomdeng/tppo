````python
from matplotlib import pyplot as plt
````

````python
import numpy as np
````

````python
ppo = np.load('Swimmer_3_results.npy')
ppo_rb = np.load('Swimmer_2_results.npy')
ppo_rb_ppo = np.load('Swimmer_1_results.npy')
````

````python
ppo_steps = ppo[0]
ppo_rewards = ppo[1]
ppo_rb_steps = ppo_rb[0]
ppo_rb_rewards = ppo_rb[1]
ppo_rb_ppo_steps = ppo_rb_ppo[0]
ppo_rb_ppo_rewards = ppo_rb_ppo[1]
````

````python
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.plot(ppo_steps, ppo_rewards, label='PPO')
plt.plot(ppo_rb_steps, ppo_rb_rewards, label='PPO_RB')
plt.plot(ppo_rb_ppo_steps, ppo_rb_ppo_rewards, label='PPO_RB PPO')
plt.xlabel('Steps')
plt.ylabel('Mean Rewards')
plt.title('Swimmer-V2')
plt.legend()
plt.savefig('swimmer_plt.png', bbox_inches='tight')
````
